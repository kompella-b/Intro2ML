```python
# ============================================================
# SETUP
# ============================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")

# Load datasets (same as your LSTM code)
def load_airline_passengers():
    df = pd.read_csv(
        "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
    )
    df["Month"] = pd.to_datetime(df["Month"])
    df.set_index("Month", inplace=True)
    return df["Passengers"].astype(float)

def load_daily_min_temps():
    df = pd.read_csv(
        "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
    )
    df["Date"] = pd.to_datetime(df["Date"])
    df.set_index("Date", inplace=True)
    return df["Temp"].astype(float)

y_air = load_airline_passengers()
y_temp = load_daily_min_temps()

print("Loaded:", len(y_air), "monthly; ", len(y_temp), "daily")
```

```python
def mae(y, yhat):
    return np.mean(np.abs(y - yhat))

def rmse(y, yhat):
    return sqrt(np.mean((y - yhat) ** 2))

def smape(y, yhat, eps=1e-8):
    return 100 * np.mean(
        2 * np.abs(yhat - y) / (np.abs(y) + np.abs(yhat) + eps)
    )

def mase(y, yhat, insample, m, eps=1e-8):
    insample = np.asarray(insample)
    if len(insample) <= m:
        m = 1
    scale = np.mean(np.abs(insample[m:] - insample[:-m]))
    if scale < eps:
        return np.nan
    return np.mean(np.abs(y - yhat)) / scale
```

```python
# ============================================================
# LIGHTWEIGHT FORECAST MODELS (FAST + NO RAM ISSUES)
# ============================================================

from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX

def fast_ets(train, h, seasonal_period):
    # Lightweight rules:
    # - Monthly: use seasonal ETS
    # - Daily: disable seasonality to avoid RAM + statsmodels bug
    if seasonal_period == 365:
        model = ExponentialSmoothing(
            train,
            trend="add",
            seasonal=None,
            initialization_method="estimated"
        ).fit(optimized=True)
    else:
        model = ExponentialSmoothing(
            train,
            trend="add",
            seasonal="add",
            seasonal_periods=seasonal_period,
            initialization_method="estimated"
        ).fit(optimized=True)

    return model.forecast(h)


def fast_arima(train, h, seasonal_period):
    # No gridsearch — fixed models
    if seasonal_period == 365:
        # Daily temps: fast model
        order = (2,0,2)
        seasonal_order = (0,0,0,0)
    else:
        # Monthly airline
        order = (1,1,1)
        seasonal_order = (1,1,1, seasonal_period)

    model = SARIMAX(
        train,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    ).fit(disp=False)

    return model.forecast(h)
```

```python
def rolling_origin_forecast(values, horizons, seasonal_period,
                            initial_train_size, origin_step):
    y = np.asarray(values, dtype=np.float32)
    N = len(y)
    max_h = max(horizons)
    insample = y[:initial_train_size]

    results = []

    for t in range(initial_train_size, N - max_h + 1, origin_step):
        train = y[:t]
        future = y[t:t + max_h]

        # ==== Fast ETS ====
        try:
            ets_pred = fast_ets(train, max_h, seasonal_period)
        except:
            ets_pred = np.full(max_h, np.nan)

        # ==== Fast ARIMA ====
        try:
            arima_pred = fast_arima(train, max_h, seasonal_period)
        except:
            arima_pred = np.full(max_h, np.nan)

        # === Compute metrics (per horizon)
        for model_name, preds in [
            ("ETS", ets_pred),
            ("ARIMA", arima_pred)
        ]:
            for h in horizons:
                y_true = future[:h]
                y_pred = preds[:h]

                results.append({
                    "model": model_name,
                    "horizon": h,
                    "MAE": mae(y_true, y_pred),
                    "RMSE": rmse(y_true, y_pred),
                    "sMAPE": smape(y_true, y_pred),
                    "MASE": mase(y_true, y_pred, insample, m=seasonal_period)
                })

    return pd.DataFrame(results)
```

```python
# ============================================================
# 1. RUN ETS + ARIMA EXPERIMENTS (baseline-style)
# ============================================================

def run_airline_baseline_ets_arima(
    y_series,
    horizons=[1, 6, 12],
    seasonal_period=12,
    initial_train_size=60,
    origin_step=1,
):
    df = rolling_origin_forecast(
        y_series.values.astype(np.float32),
        horizons=horizons,
        seasonal_period=seasonal_period,
        initial_train_size=initial_train_size,
        origin_step=origin_step
    )

    if df.empty:
        return df

    df["dataset"] = "airline"
    df["use_calendar"] = False
    df["window"] = "N/A"      # no window concept for ETS/ARIMA
    return df


def run_temps_baseline_ets_arima(
    y_series,
    horizons=[1, 7, 14],
    seasonal_period=365,
    initial_train_size=365,
    origin_step=7,
):
    df = rolling_origin_forecast(
        y_series.values.astype(np.float32),
        horizons=horizons,
        seasonal_period=seasonal_period,
        initial_train_size=initial_train_size,
        origin_step=origin_step
    )

    if df.empty:
        return df

    df["dataset"] = "temps"
    df["use_calendar"] = False
    df["window"] = "N/A"
    return df
```

```python
def plot_last_year_forecast_classical(
    y_series,
    seasonal_period,
    model_name="ETS",
    horizon=12
):
    values = y_series.values.astype(np.float32)
    N = len(values)
    train_end_idx = N - horizon
    train = values[:train_end_idx]
    test  = values[train_end_idx:]

    if model_name.upper() == "ETS":
        pred = fast_ets(train, horizon, seasonal_period)
    else:
        pred = fast_arima(train, horizon, seasonal_period)

    idx = y_series.index
    hist_len = min(3*horizon, N)

    plt.figure(figsize=(10,4))
    plt.plot(idx[-hist_len:], values[-hist_len:], label="History")
    plt.plot(idx[train_end_idx:], test, label="True future", marker="o")
    plt.plot(idx[train_end_idx:], pred, label=f"{model_name} forecast", marker="x")
    plt.title(f"{model_name} last-year forecast (h={horizon})")
    plt.grid(True); plt.legend(); plt.tight_layout()
    plt.show()
```

```python
air_results = rolling_origin_forecast(
    y_air,
    horizons=[1, 6, 12],
    seasonal_period=12,
    initial_train_size=60,
    origin_step=1
)

temp_results = rolling_origin_forecast(
    y_temp,
    horizons=[1, 7, 14],
    seasonal_period=365,
    initial_train_size=365,
    origin_step=7
)
```

```python
def plot_rolling_origin_schematic(N, initial, horizon, step=12):
    x = np.arange(N)
    origins = list(range(initial, N-horizon+1, step))[:3]

    fig, ax = plt.subplots(figsize=(8,2))
    for i, o in enumerate(origins):
        ax.axvspan(0, o, alpha=0.1, color="blue")
        ax.axvspan(o, o+horizon, alpha=0.3, color="orange")
        ax.text(o + horizon/2, 0.1*(i+1),
                f"Origin {i+1}", ha="center")

    ax.set_yticks([])
    ax.set_title("Rolling-Origin Backtesting Schematic")
    plt.show()

plot_rolling_origin_schematic(len(y_air), initial=60, horizon=12)
```

```python
print("Running Airline ETS+ARIMA baseline...")
airline_baseline = run_airline_baseline_ets_arima(y_air)
display(
    airline_baseline
    .sort_values(["model", "horizon"])
    .reset_index(drop=True)
)

print("\nRunning Temps ETS+ARIMA baseline...")
temps_baseline = run_temps_baseline_ets_arima(y_temp)
display(
    temps_baseline
    .sort_values(["model", "horizon"])
    .reset_index(drop=True)
)
```

```python
def plot_rolling_origin_schematic(
    series_length,
    initial_train_size,
    horizon,
    origin_step=1,
    num_origins=3
):
    """
    Simple schematic showing expanding-window origins and test blocks.
    """

    N = series_length
    origins = list(range(initial_train_size, N - horizon + 1, origin_step))[:num_origins]

    import numpy as np
    import matplotlib.pyplot as plt

    fig, ax = plt.subplots(figsize=(8, 2))
    x = np.arange(N)
    ax.plot(x, np.zeros_like(x), alpha=0)

    for i, origin in enumerate(origins):
        ax.axvspan(0, origin, alpha=0.1, color="tab:blue")
        ax.axvspan(origin, origin + horizon, alpha=0.3, color="tab:orange")
        ax.text(origin + horizon / 2, 0.02 * (i + 1),
                f"Origin {i+1}", ha="center", va="bottom")

    ax.set_yticks([])
    ax.set_xlabel("Time index")
    ax.set_title("Rolling-origin (expanding-window) schematic")
    plt.tight_layout()
    plt.show()
```

```python
# Airline
plot_last_year_forecast_classical(y_air, 12, model_name="ETS", horizon=12)
plot_last_year_forecast_classical(y_air, 12, model_name="ARIMA", horizon=12)

# Daily Temps
plot_last_year_forecast_classical(y_temp, 365, model_name="ETS", horizon=14)
plot_last_year_forecast_classical(y_temp, 365, model_name="ARIMA", horizon=14)
```
